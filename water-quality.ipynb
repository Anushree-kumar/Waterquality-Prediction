```python
multi = train["result"].value_counts() > 1
filtered = train[train["result"].isin(multi[multi].index)]
train = filtered
train
```

```python
train = train.dropna()
train
```

```python
target = train['result']

ds= train.drop(['result'], axis=1).append(test)
ds
```

```python
ds.drop(['id'],axis=True)
```

```python
for col in ds:
    ds[col].fillna(0, inplace=True)
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()

for col in ds:
    if ds[col].dtype=="object":
        ds[col] = enc.fit_transform(ds[col].values.reshape(-1,1))
ds
```

```python
corr = ds.corr()
```

```python
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
select = ds.columns[columns]
ds = ds[select]
ds
```

```python
Y = target
X = ds[: len(train)]
X_test = ds[len(train) :]
```

```python
X['target'] = target

train = X.sample(frac=.9, random_state=42)
val = X.drop(train.index)

X_train = train.iloc[:, :-1]
Y_train = train.iloc[:,-1]

X_val = val.iloc[:, :-1]
Y_val = val.iloc[:,-1]

X_train.shape, Y_train.shape, X_val.shape, Y_val.shape, X_test.shape
```

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(random_state=42).fit(X_train, Y_train)
print(model.score(X_train, Y_train))
```

```python
pred = model.predict(X_val)
print(model.score(X_val, Y_val))
```

```python
final=pd.DataFrame({'Actual': Y_val, 'Predicted':pred})
final
```

```python
Prediction = model.predict(X_test)
Prediction
```

```python
sub['result'] = Prediction
sub.to_csv('submission.csv', index=False)
sub = pd.read_csv("submission.csv")
```
